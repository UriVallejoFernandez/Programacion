sumar tots aquest valors numerics amb SUM cal que la sumi digui **l'import total és...**

```python
import pandas as pd

numerica = pd.Series([12, 58, 63, 47, 81, 35])

total = numerica.sum()
print("L'import total és",total)
```

Fer una mitjana

```python
import pandas as pd

numerica = pd.Series([12, 58, 63, 47, 81, 35])

total = numerica.sum()
mitjana = numerica.mean()

print("L'import total és", total)
print("La mitjana és", mitjana)
```

Redondear a 2 decimals

```python
import pandas as pd

numerica = pd.Series([12, 58, 63, 47, 81, 35])

total = numerica.sum()
mitjana = round(numerica.mean(),2)



print("L'import total és", total)
print("La mitjana és", mitjana)
```

```python
import pandas as pd

numerica = pd.Series([12, 58, 63, 47, 81, 35])

minim = numerica.min()

print("El minim és", minim)
```

la suma acumulada de valor .cumsum

```python
import pandas as pd

numerica = pd.Series([12, 58, 63, 47, 81, 35])

total = numerica.cumsum()
print(total)
```

comptar la quantitat de cada valor (value_counts)

```python
import pandas as pd

numerica = pd.Series([12, 58, 63, 47, 81, 35])

total = numerica.value_counts()
print(total)
```
