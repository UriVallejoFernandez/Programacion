```python
import pandas as pd

la_liga = pd.DataFrame({
    'Equip': ['Barça', 'Madrid', 'Atletico de Madrid', 'Athletic', 'Villareal', 'Mallorca'],
    'Partits jugats': [16, 14, 15, 15, 14, 16],
    'Punts': [37, 33, 32, 26, 26, 24],
    'Gols a favors': [48, 30, 26, 22, 27, 16]
})

la_liga["Gols en contra"] = [17, 13, 8, 15, 23, 18]

promig_gols_a_favors = la_liga["Gols a favors"].mean().round(2)
promig_punts = la_liga["Punts"].mean().round(2)

primer = la_liga[la_liga["Punts"] == la_liga["Punts"].max()]
ultim = la_liga[la_liga["Punts"] == la_liga["Punts"].min()]

print("El promig de gols és de", promig_gols_a_favors)
print("El promig de punts és de", promig_punts)
print("L'equip amb més punts és:", primer['Equip'].values[0])
print("L'equip amb menys punts és:", ultim['Equip'].values[0])

la_liga
```
